### Step 1. Import the necessary libraries

```python
from pathlib import Path
import pandas as pd


def data_path(filename):
    matches = list(Path.cwd().rglob(filename))
    return matches[0] if matches else filename
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo.item_price.dtype
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo.item_price = chipo.item_price.str.replace("$", "", regex=False).astype(float)
chipo.head()
```

#### Step 13.c. Check the item price type

```python
chipo.item_price.dtype
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo.quantity * chipo.item_price).sum()
revenue
```

### Step 15. How many orders were made in the period?

```python
chipo.order_id.nunique()
```

### Step 16. What is the average revenue amount per order?

```python
order_revenue = chipo.groupby("order_id").apply(lambda order: (order.quantity * order.item_price).sum())
order_revenue.mean()
```

```python
chipo["revenue"] = chipo.quantity * chipo.item_price
chipo.groupby("order_id").revenue.sum().mean()
```
